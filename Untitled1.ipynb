```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt



image = cv2.imread(r'c:\Users\HANU\Desktop\img2.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gradX = cv2.Sobel(gray, ddepth = cv2.CV_32F, dx = 1, dy = 0, ksize = -1)
gradY = cv2.Sobel(gray, ddepth = cv2.CV_32F, dx = 0, dy = 1, ksize = -1)

#sobel x로 수직선 방향의 에지를 검출
#sobel y로 수평 방향의 에지를 검출

#sobel은 Gaussian Smoothing과 미분연산을 결합한 형태의 연산을 수행함
#ddepth:결과 이미지 데이터 타입(이미지 픽셀값을 float32로 설정)
#dx,dy:각각 x,y방향으로 미분 차수( 1,0이면 x방향으로 1차 미분수행 y는 그대로)
#ksize:확장 sobel커널의 크기로 -1일때는 scharr 필터를 적용해 좀더 정확한 결과가 나옴
gradient = cv2.subtract(gradX, gradY)
#두개의 이미지를 -연산한다
gradient = cv2.convertScaleAbs(gradient)
#결과에 절대값을 적용한다

plt.figure(figsize=(10,10))
plt.subplot(221),plt.imshow(gradX, cmap='gray'),plt.title('sobelX')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(gradY, cmap='gray'),plt.title('sobelY')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(gradient,cmap='gray'),plt.title('convertScaleAbs')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blurred = cv2.GaussianBlur(gradient, (9,9),0)
(_, thresh) = cv2.threshold(blurred, 150, 255, cv2.THRESH_BINARY)
#cv2.threshold(img,threshold_value,value,flag) 임계값 이상이면 어떤 값으로 바꾸어주고 낮으면 0으로 바꾼다
#threshold_value:하나의 이미지에 전역으로 적용될 하나의 픽셀 임계값
#value:픽셀 임계값보다 클때 적용되는 최대값
#flag:임계값 적용방법 cv2.thresh_binary:픽셀값이 threshold_value보다 크면 value,작으면 0으로 할당


closed = cv2.erode(thresh, None, iterations = 4)
closed = cv2.dilate(thresh, None, iterations = 4)

#4번의 이미지 침식과 4번의 이미지 팽창을 수행
#작은 얼룩을 제거하고 나머지 흰색 픽셀을 확장
plt.figure(figsize=(10,10))
plt.subplot(221),plt.imshow(blurred, cmap='gray'),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(thresh, cmap='gray'),plt.title('threshold')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(closed,cmap='gray'),plt.title('closed')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cnts, hierarchy = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#여기실행안되면 cnts 앞에 _,붙이세요
c = sorted(cnts, key = cv2.contourArea, reverse = True)[0]

#cv2.findContours: 이미지에서 찾는 contours와 그의 계층구조를 반환한다
#cv2.RETR_TREE:모든 contour을 추출하고 contour들간의 상관관계를 추출한다
#cv2.CHAIN_APPROX_SIMPLE:contour 근사방법으로 contour의 수평,수직,대각선 방향의 점을 버리고 끝점만 남긴다
#컨튜어를 정렬한다


# compute the rotated bounding box of the largest contour
#가장 큰 contour의 박스를 계산한다
rect = cv2.minAreaRect(c)
#인자로 입력한 contour에 외접하면서 면적이 가장 작은 직사각형을 구한다
box = np.int0(cv2.boxPoints(rect))
#boxpoints로 꼭지점4개의 좌표를 얻고 정수값으로 전환한다

```

```python
cv2.drawContours(image, [box], -1, (0, 255, 0), 2)

#[box]:img에 그릴값
#-1:실제로 그릴 contours 음수이면 모든 contours를 그린다

plt.figure(figsize=(10,10))
plt.subplot(121),plt.imshow(image, cmap='gray'),plt.title('finalimage')
plt.xticks([]), plt.yticks([])
#plt.subplot(122),plt.imshow(img_crop, cmap='gray'),plt.title('img_crop')
#plt.xticks([]), plt.yticks([])

#plt.show()
#cv2.imshow("image",image)


#cv2_imshow(gradient)
#cv2_imshow(blurred)
#cv2_imshow(thresh)
#cv2_imshow(closed)

img_crop = image[box[2][1]:box[0][1], box[1][0]:box[3][0]]
cv2.imshow("img_crop",img_crop)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite("img_crop.jpg",img_crop)
```
